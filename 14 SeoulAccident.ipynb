# 서울시 교통사고 데이터

```r
seoul <- read.csv("../Data/newSeoulAccident.csv")
# 데이터 소개
# 서울 교통사고 데이터는 연도별 월별 구별로 정리된 데이터 이며 각 구분자에 발생건수, 사망자수, 부상자수
# 데이터를 포함하고 있다.
# 총 2011년 1월 부터 2017년 12월까지 정리 된 데이터 이다.
seoul




```

## 자치구별 발생건수의 변동계수 구하기

```r
# 자치구별 발생건수 변동계수를 구하여 교통사고 다발 자치구 Top5 출력
sigu.m <- aggregate(발생건수 ~ 자치구명, seoul, mean)
sigu.sd <- aggregate(발생건수 ~ 자치구명, seoul, sd)
sigu.m.sd<- merge(sigu.m,sigu.sd, by = "자치구명")



```

```r
sigu.m.sd$cv <- round(sigu.m.sd$발생건수.y /  sigu.m.sd$발생건수.x ,3)



```

```r
library("doBy")



```

```r
orderBy.m.sd.cv<-orderBy(~ cv , sigu.m.sd)
names(orderBy.m.sd.cv) <- c('자치구명', 'mean' ,'sd' ,'cv')
orderBy.m.sd.cv



```

## 자치구별 사망건수의 변동계수 구하기

```r
# 자치구별 사망건수 변동계수를 구하여 교통사고 사방 다발 자치구 출력
sigu.m <- aggregate(사망자수 ~ 자치구명, seoul, mean)
sigu.sd <- aggregate(사망자수 ~ 자치구명, seoul, sd)
sigu.m.sd<- merge(sigu.m,sigu.sd, by = "자치구명")



```

```r
sigu.m.sd$cv <- round(sigu.m.sd$사망자수.y /  sigu.m.sd$사망자수.x ,3)



```

```r
orderBy.m.sd.cv<-orderBy(~ cv , sigu.m.sd)
names(orderBy.m.sd.cv) <- c('자치구명', 'mean' ,'sd' ,'cv')
orderBy.m.sd.cv



```

## 자치구별 부상자수의 변동계수 구하기

```r
# 자치구별 부상자수의 변동계수를 구하여 교통사고 사방 다발 자치구 출력
sigu.m <- aggregate(부상자수 ~ 자치구명, seoul, mean)
sigu.sd <- aggregate(부상자수 ~ 자치구명, seoul, sd)
sigu.m.sd<- merge(sigu.m,sigu.sd, by = "자치구명")



```

```r
sigu.m.sd$cv <- round(sigu.m.sd$부상자수.y /  sigu.m.sd$부상자수.x ,3)

orderBy.m.sd.cv<-orderBy(~ cv , sigu.m.sd)
names(orderBy.m.sd.cv) <- c('자치구명', 'mean' ,'sd' ,'cv')
orderBy.m.sd.cv



```

```r
# 2015년 12월달의 data 추출

traffice2015_12 <- subset(seoul, 연도==2015 & 월==12)
traffice2015_12



```

```r
gp <- ggplot(data = traffice2015_12, aes(x=발생건수, y=부상자수))
gp <- gp + scale_fill_brewer(palette = "Set1")
gp <- gp + geom_point(aes(size=사망자수),shape = "★", color = "red",
                     alpha=0.55)
gp <- gp + ggtitle("교통사고 발생현황에 따른 버블 차트(2015.12)")
print(gp)







```

```r
# 교통사고 발생건수와 부상자수의 상관관계, 추세선
ab <- ggplot(data = traffice2015_12, aes(x=발생건수, y=부상자수))
ab <- ab + scale_fill_brewer(palette = "Set1")
ab <- ab + geom_point()
ab <- ab + ggtitle("교통사고 발생건수와 부상자수의 관계(2015.12)")
ab <- ab + stat_smooth(method = "lm", se = T, color = "red")
print(ab)







```
