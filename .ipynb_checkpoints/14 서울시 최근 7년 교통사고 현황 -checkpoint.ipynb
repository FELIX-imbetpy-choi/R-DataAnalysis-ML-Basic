```r
seoul<- read.csv('../Data/newSeoulAccident.csv')
```

```r
head(seoul)
```

```r
year_acc<-aggregate(발생건수 ~연도 ,seoul, sum)
year_death<-aggregate(사망자수 ~연도 ,seoul, sum)
year_sick<-aggregate(부상자수 ~연도 ,seoul, sum)
```

```r
year_all <- cbind(year_acc,year_death,year_sick)
```

```r
year_all <- year_all[,c(-3,-5)]
```

# 서울시 최근 7년 교통사고 현황

```r
plot(year_all[1,], 
     type='o',
     col ='red', # 칼라
    ylim=c(0,60000),
     axes =F, # 테두리 표시 여부 
     ann =F # x,y 축 제목 표시여부    
    )

# plot은 하나 있어야하구 그위에 레이어를 쌓아서 보는것임 
# x축 y 축의 라벨을 설정
axis(1, # x축
    at =1:7, # 데이터 범위
    lab =c('2011','2012','2013','2014','2015','2016','2017'), # 레이블 이름 
    )
axis(2, # y축
    ylim=c(0,50000)
    )
# Main TIele 지정
title(main ="서울시 최근 7년간 연도별 교통사고 추이", # 제목
      col.main ='orange', # 제목 색깔
      font.main=4 # font type
     )
# X축 타이츨 
title(xlab='년도',
     col.lab ='black' # x 축 제목 색깔
     )
title(ylab='count',
     col.lab ='black' # x 축 제목 색깔
     )
# 선그리기 같은 칸에 코드를 적어줘야햄
lines(year_all[2,]*100, type = 'o',pch =21 ,col='green',lty =2)
lines(year_all[3,], type = 'o',pch =22 ,col='blue',lty =3)

# legend(범례 그리기)
legend(6,20000, # 좌표위치 범례가 들어갈 위치 ㅎ
       c('발생건수','사망자수*100','부상자수'), # data칼럼
       cex =0.75, # 1을 기준으로 한 상대적 크기
        col =c('red','green','blue'), # 색깔
       pch =c(21,21,22),
       lty = 1:3
      )
```

```r
seoul_year_sick<-aggregate(발생건수 ~ 연도+자치구명 ,seoul, sum)

seoul_year_sick_2011<- seoul_year_sick[seoul_year_sick[,1]==2011,]
seoul_year_sick_2012<- seoul_year_sick[seoul_year_sick[,1]==2012,]
seoul_year_sick_2013<- seoul_year_sick[seoul_year_sick[,1]==2013,]
seoul_year_sick_2014<- seoul_year_sick[seoul_year_sick[,1]==2014,]
seoul_year_sick_2015<- seoul_year_sick[seoul_year_sick[,1]==2015,]
seoul_year_sick_2016<- seoul_year_sick[seoul_year_sick[,1]==2016,]
seoul_year_sick_2017<- seoul_year_sick[seoul_year_sick[,1]==2017,]

row.names(seoul_year_sick_2011) =seoul_year_sick_2011[,2]
year_sick_2011<-seoul_year_sick_2011[,-2]

row.names(seoul_year_sick_2012) =seoul_year_sick_2012[,2]
year_sick_2012<-seoul_year_sick_2012[,-2]


row.names(seoul_year_sick_2013) =seoul_year_sick_2013[,2]
year_sick_2013<-seoul_year_sick_2013[,-2]


row.names(seoul_year_sick_2014) =seoul_year_sick_2014[,2]
year_sick_2014<-seoul_year_sick_2014[,-2]


row.names(seoul_year_sick_2015) =seoul_year_sick_2015[,2]
year_sick_2015<-seoul_year_sick_2015[,-2]


row.names(seoul_year_sick_2016) =seoul_year_sick_2016[,2]
year_sick_2016<-seoul_year_sick_2016[,-2]


row.names(seoul_year_sick_2017) =seoul_year_sick_2017[,2]
year_sick_2017<-seoul_year_sick_2017[,-2]


sick_year<- cbind(year_sick_2011,year_sick_2012,year_sick_2013,year_sick_2014,year_sick_2015,year_sick_2016,year_sick_2017)
sick_year_all_balsang<-sick_year[,c(2,4,6,8,10,12,14)]

names(sick_year_all_balsang)<-c(2011,2012,2013,2014,2015,2016,2017)

t(sick_year_all_balsang)
sick_year_all_balsang<-as.matrix(sick_year_all_balsang)
sick_year_all_balsang <- t(sick_year_all_balsang)
row.names(sick_year_all_balsang) <- c("11","12","13","14","15","16","17")

sick_year_all_balsang<-t(sick_year_all_balsang)
sick_year_all_balsang_15_17<-sick_year_all_balsang[,c(5,7)]

balsang_bar<-sick_year_all_balsang_15_17[,1] - sick_year_all_balsang_15_17[,2]
balsang_bar<-sort(balsang_bar,decreasing = T)
balsang_bar
```
