```r
# Baby names 읽어서 head 보기
baby<-read.csv("../Data/R_Baby_Names_1880-2014.csv")
head(baby)




```

```r
nrow(baby)



```

```r
# 데이터 정제하기
# year 중에서 연도 정보만 빼오기
baby$Year <- substr(baby$Year, 7, 10)  # 7번째 값에서 10번째 값까지 쓰기
head(baby)



```

```r
# data 분류
# 성별이 F인 데이터만 보아서 Dataset 이름 FF로 만들기
FF <- subset(baby, baby$Gender== "F")
MM <- subset(baby, baby$Gender== "M")



```

```r
Babycount <- aggregate(Count ~ Year -Gender , baby, sum) 
head(Babycount)



```

```r
Fcount<-Babycount[Babycount$Gender == "F",c(1,3)]
Mcount<-Babycount[Babycount$Gender == "M",c(1,3)]
baby_merge<-merge(Mcount,Fcount, by="Year")
colnames(baby_merge) <- c('Year', 'M', 'F')
baby_merge







```

```r
# 1번째 선긋기
plot(x=baby_merge$Year, y=baby_merge$M , main="연도별 아가 탄생", type="l", col="blue",ylim=c(0,3000000), xlab = "연도", ylab = "아가수")
# 2번째 선긋기
lines(x=baby_merge$Year, y=baby_merge$F, pch = 21, type="l" ,col ="red")

# legend(범례)
legend(2000,3000000, # 범례 좌표위치
       c("M","F"), # data 컬럼
       cex = 1, # 1을 기준으로 한 범례의 상대적 크기
       col = c("blue","red"), #색깔
       pch=c("★","★")
      )







```
