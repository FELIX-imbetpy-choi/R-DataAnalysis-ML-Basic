# 반복문

```r
# 반복문
sum1 <- 0
for(i in seq(1,10,1)){
    sum1 <- sum1+i
}
sum1




```

```r
# 반복문 연습
dan <- 3
for (i in seq(1,9)){
    print(paste(dan,"X",i,"=",dan*i))
}



```

```r
gugu <- function(){
    dan <- readline("단을 입력하세요") # 사용자로 부터 입력 받는 함수
    dan <- as.integer(dan)
    for (i in seq(1,9)){
        print(paste(dan,"X",i,"=",dan*i))
    }
}



```

```r
gugu()



```

```r
gugu1 <- function(){
    for (i in seq(2,9)){
        print(paste(rep("=",i))
        for (k in seq(1,9)){
            print(paste(i,"X",k,"=",k*i))
        }
    }
}







```

```r
gugu1()



```

```r
gugu2 <- function(){
    for (i in seq(2,9)){
        cat(paste(" ",i,"X",seq(1,9),"=",seq(1,9)*i,"\n"))
        cat(paste('------','\n'))
}
    }







```

```r
gugu3 <- function(){
    for (i in seq(2,9)){
        cat(paste("\t",i,"X",seq(1,9),"=",seq(1,9)*i,"\n"))
        cat(paste('---------------------------------',"\n"))
    }
}







```

```r
gugu3()



```

### 번외 반복문

```r
# 반복문 while
sum1 <- 0
i <- 1
while(i<=10){
    sum1 <- sum1 + i
    i <- i + 1
}
sum1







```

```r
# 반복문 repeat
sum1 <- 0
i <- 1
repeat{
    sum1 <- sum1+i
    i <- i+1
    if(i > 10)break
    
}
sum1







```

# 예제

#### 시험점수가 78, 95, 85, 65 일 경우 
#### 80점 이상이면 우수 , 그렇지 않으면 노력이라고 출력하시오   
#### (단 if문을 사용하지 말고 다른 함수를 사용)

```r
score1=data.frame(x=c(78, 95, 85, 65))
judg <- function(){
    kk<-subset(score1, x>=80)
    cat(print("우수"))
    print(kk)
    jj<-subset(score1, x<80)
    cat(print("노력"))
    print(jj)
}







```

```r
judg()



```

```r
score <- c(78,95,85,65)
df = data.frame(score=score,결과=ifelse(score >= 80, '우수', '노력'))
df



```
