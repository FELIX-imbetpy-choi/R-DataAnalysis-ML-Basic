# 변수와 자료형
## 변수
분석에 필요한 자료를 일시적으로 저장하거나 처리 결과를 담을 수 있는 기억장소를 지정해 주는 역할.   
변수 이름은 값을 저장하는 메모리 영역의 이름으로 할당.   


### ※ 변수 이름의 작성 규칙
- 첫자는 영문으로 시작
- 첫자를 제외하고 숫자, uder bar, dot 등을 사용할 수 있다
- 대문자 소문자 구분
- 변수이름은 의미를 파악할 수 있는 이름으로 저장하는 것이 좋다
- 두단어를 포함하여 변수 이름을 지정할 경우 두번째 단어의 첫문자는 대문자로 표시하거나 언더바로 표시하는 것이 좋다
- 한번 정의된 변수는 재사용이 가능하고, 가장 최근에 할당된 값으로 수정된다

## 자료형
숫자 : 정수, 실수   
문자 : 문자, 문자열   
논리 : 참, 거짓   
결측치 : Na(Not Available), NaN(Not a Number )

```r
int <- 20
int
string <- '홍길동'
string
boolean <- 'T'
boolean
sum(10,20,30) # 3개 숫자의 합
sum(10,20,30,NA) # 결측치가 있는 숫자의 합, 결측치 제거후 합산
sum(10,20,30,NA,na.rm = T) # 결측치 제거 후 합산







```

```r
## 자료형 확인
is.character(string)
is.numeric(int)
s <- is.logical(boolean)
is.na(s)
y <- c(1,2,3,NA)
is.na(y)





```

```r
# 자료형 변환
x <- c(1,2,'3')
x
# sum(x) # 벡터를 확인해 보고 계산을 해야한다.
x1 <- as.numeric(x)
sum(x1)



```

## 요인형 변환
요인(factor)은 동일한 값의 목록을 범주(category)로 갖는 벡터자료   
범주는 변수가 가질 수 있는 값의 범위(예 : 성별 변수의 범주는 남자와 여자)

```r
str(iris)

```

```r
gender <- c('man','woman','woman','man','man')
gender
# plot(gender) # 문자기 때문에 오류가 발생한다. 이 때 요인형 변환 -> 차트를 만들 수 있음
Ngender <- as.factor(gender)
Ngender
str(gender)
str(Ngender) # 구성 살펴보기 - factor 로 되어 있음
plot(Ngender)





```

```r
table(Ngender) #알아서 빈도수를 알려주는 함수
table(gender)



```

```r
# 날짜형 변환
as.Date('17/08/08','%y/%m/%d') # 년도가 2자리인 경우
as.Date('17-08-08','%y-%m-%d')

```

```r
as.Date('2017/08/08','%Y/%m/%d') # 년도가 4자리인 경우

```

```r
# 여러개의 날짜 데이터 변환
datas <- c('02/28/17','02/29/17','03/01/17') # 17년 2월 29일은 없다. 없는 날짜는 변경할 수 없음
as.Date(datas,'%m/%d/%y')

```
