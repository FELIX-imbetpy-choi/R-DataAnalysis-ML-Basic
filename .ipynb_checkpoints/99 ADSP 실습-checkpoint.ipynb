```r
# 상관분석 예제 p305 
data(mtcars)
a <- mtcars$mpg
b <- mtcars$hp

cor(a,b) # 상관관계 --> -0.77
cov(a,b) # 분산 --> -320.73
cor.test(a, b, method = 'pearson') # cor -0.77

# 결과 및 해석
```

```r
# 단순선형회귀분석 - p 311 
# 에어컨 예약 대수와 판매대수를 회귀분석
x <- c(19, 23, 26, 29, 30, 38, 39, 46, 49)
y <- c(33, 51, 40, 49, 50, 69, 70, 64, 89)
lm(y~x)
summary(lm(y~x))

# 결과 해석하기
# 회귀식에 대한 정보를 살펴면 p-value 가 0.005805 로 0.05보다 낮아 회귀계수의 추정치들이 통계적으로 유의미하다고 볼 수 있다.
# 따라서 귀무가설을 기각하고 대립가설을 채택한다. 따라서 예약대수와 판매대수는 관련이 있다.
# 분석에 대한 설명력은 R - squared 0.8341 로 높게 나타남으로 이 회귀식이 데이터를 적절하게 설명하고 있다고 볼 수 있다.
```

```r
# 다중회귀분석 - p314
library(MASS)
head(Cars93)
attach(Cars93)
lm(Price~EngineSize+RPM+Weight, data = Cars93) # 회귀식 작성 엔진사이즈 / 알피엠 / 무게 에 따라 가격은 어떻게 달라지는가?
summary(lm(Price~EngineSize+RPM+Weight, data = Cars93))

# 결과 및 해석
# p-value 값이 유의 수준이므로 추정된 회귀모형이 통계적으로 유의함을 알 수 있다.
# R - squared (결정계수)가 낮아 데이터의 설명력은 낮지만/ p-value에 따라 회귀분석 결과에서 회귀식과 회귀계수이 통계적으로 유의하므로
# 자동차의 가격을 엔진크기과 알피엠 무게로 추정이 가능하다.!!!
```
