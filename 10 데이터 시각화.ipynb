# 선거 데이터

```r
election <- read.csv("..//Data//election_result_ansi.csv")
head(election)



```

#### 선거데이터에서 pop는 전체 주민 수 이므로 세후보의 득표수 합이 pop와 같을 수 없다.
#### 먼저 세 후보의 득표수 전체를 각 지역별로 산출해 보자.

```r
# election<-cbind(election,electionSum=apply(election[5:7],1,sum))
summha<-apply(election[5:7],1,sum)
election$electionSum=summha
head(election)



```

#### electionSum에서  세후보의 득표비율을 구하여 electionRatio 라는 컬럼이름으로 추가하기

```r
# electionSum의 선거비율을 구하여 electionRatio 라는 컬럼이름으로 electionRatio에 추가하기
election<-cbind(election,electionRatio=round(election$electionSum/election$pop*100,2))
head(election)



```

### 선거비율 함수 만들기

```r
# 선거비율 함수 만들기
ratioF <- function(x,output){
    round(as.integer(x[8])/as.integer(x[4])*100, 2)
}



```

```r
electionRatio <- apply(election, 1, ratioF)
election <- cbind(election, electionRatio)
head(election)



```

## 시각화1) 각 광역시 별 세후보의 광열시별 득표율을 파이차트화
#### 각 광역시 별 electionSum / pop * 100

```r
head(election)



```

```r
# 각 시도별 전체 득표수 , 세 후보 득표수 만들어 합치기
# 광역시별 전체, 세후보 득표수 만들기
get_pop <- aggregate(pop ~ 광역시도 , election, sum) 
get_electionSum <- aggregate(electionSum ~ 광역시도 , election, sum) 
# 데이터 합치기
by_merge<-merge(get_pop,get_electionSum, by="광역시도")
# 비율 만들기
vote_rate_three<-round(by_merge[3]/apply(by_merge[2],2,sum)*100,2)
colnames(vote_rate_three)<-c("vote_rate_three") # 굳이 컬럼 이름 바꾸기
# 붙이기
complete_election<-cbind(by_merge,vote_rate_three)
# 그림 그릴 부분만 떼어내기
complete_election_for_chart<-complete_election[-2:-3]
# 내림차순으로 정렬
complete_election_for_chart<-orderBy(~ -vote_rate_three,complete_election_for_chart)
complete_election_for_chart







```

```r
label <- sprintf("%s \n %.1f",complete_election_for_chart$광역시도,complete_election_for_chart$vote_rate_three)
pie3D(complete_election_for_chart$vote_rate_three, main ='광역시별 투표율',labels =label,labelcex = 0.5)



```

### 선생님 진행

```r
# 각 후보의 득표수 만들기
elec_moon<-aggregate(moon ~ 광역시도, election, sum) 
elec_hong<-aggregate(hong ~ 광역시도, election, sum) 
elec_ahn<-aggregate(ahn ~ 광역시도, election, sum) 



```

```r
# 투표인이 가장 많은 4지역 구하기
elec_sum<-aggregate(pop ~ 광역시도, election, sum) 
top4Area <- orderBy(~ -pop, elec_sum)[1:4,1]
top4Area
# 지역 1순위의 각 후보별 투표수
top4_1 <- c(subset(elec_moon, 광역시도 == top4Area[1])[2],
            subset(elec_hong, 광역시도 == top4Area[1])[2],
            subset(elec_ahn, 광역시도 == top4Area[1])[2]
           )
# 지역 2순위의 각 후보별 투표수
top4_2 <- c(subset(elec_moon, 광역시도 == top4Area[2])[2],
            subset(elec_hong, 광역시도 == top4Area[2])[2],
            subset(elec_ahn, 광역시도 == top4Area[2])[2]
           )
# 지역 3순위의 각 후보별 투표수
top4_3 <- c(subset(elec_moon, 광역시도 == top4Area[3])[2],
            subset(elec_hong, 광역시도 == top4Area[3])[2],
            subset(elec_ahn, 광역시도 == top4Area[3])[2]
           )
# 지역 4순위의 각 후보별 투표수
top4_4 <- c(subset(elec_moon, 광역시도 == top4Area[4])[2],
            subset(elec_hong, 광역시도 == top4Area[4])[2],
            subset(elec_ahn, 광역시도 == top4Area[4])[2]
           )











```

```r
top4_1d <- as.data.frame(top4_1)
top4_2d <- as.data.frame(top4_2)
top4_3d <- as.data.frame(top4_3)
top4_4d <- as.data.frame(top4_4)



```

```r
# 상자수염 그래프(boxplot) - 분석가들이 많이 사용하는 것
# 시즌별 Baseball 판매 현황
abc <- c(110, 330, 150, 280, 310)
# 시즌별 soccerball 판매 현황
def <- c(180, 200, 210, 190, 170)
# 시즌별 beachball 판매 현황
ghi <- c(210, 150, 160, 210, 70)







```

```r
boxplot(abc,def,ghi)

# 이상치를 살펴보려고 그린다
# 중앙값에 선이 그어져 있는데 이 선은 상위부터 순서상 중간값이다.
# 중앙값이 높이 있다는 것은 판매량이 계속 높은 구간이 많다는 뜻이다.







```

```r
# 때에 따라 상자수염 그래프를 옆으로 보는것이 좋을 때가 있다.
boxplot(abc,def,ghi, horizontal = T,
       col = c("yellow","cyan","green"),
       names = c("Baseball","soccerball","beachball"))



```

```r
jkl <- c(1,1,1,1,1,2,5)
summary(jkl)
# 4번째 값이 중앙값으로 보인다.
boxplot(jkl)
# 5 가 이상치로 떨어져 나온다 평균값과 표준편차 계산으로 이상치를 분리해 낸다.



```

```r
data(galton) # 부모와 자식간의 키의 관계를 나타낸 데이터, 단위 : 인치
head(galton)

# 상관관계
cor(galton$parent,galton$child)
str(galton)



```

```r
# 빈도수를 데이터 프레임으로 변경
table(galton$parent,galton$child)



```

```r
galtonData <- as.data.frame(table(galton$child,galton$parent))
str(galtonData)
names(galtonData) <- c("child","parent","Freq")
head(galtonData)



```

```r
parent <- as.numeric(galtonData$parent)
child <- as.numeric(galtonData$child)



```

```r
plot(parent,child, pch=21,
    col = "blue", bg = "green",
    cex = 0.2 * galtonData$Freq,
    xlab = "parent",
    ylab = "child")



```

### 3차원 산점도 시각화
install.packages("scatterplot3d")

```r
library(scatterplot3d)



```

```r
# 꽃의 종류별 분류
data(iris)
iris_setosa <- iris[iris$Species == "setosa",]
iris_virginica <- iris[iris$Species == "virginica",]
iris_versicolor <- iris[iris$Species == "versicolor",]



```

```r
# 3차원 틀 생성
d3 <- scatterplot3d(iris$Petal.Length, iris$Sepal.Length, iris$Sepal.Width)

d3$points3d(iris_setosa$Petal.Length,iris_setosa$Sepal.Length,
           iris_setosa$Sepal.Width, bg = "orange", pch = 21)

d3$points3d(iris_virginica$Petal.Length,iris_virginica$Sepal.Length,
           iris_virginica$Sepal.Width, bg = "blue", pch = 24)

d3$points3d(iris_versicolor$Petal.Length,iris_versicolor$Sepal.Length,
           iris_versicolor$Sepal.Width, bg = "red", pch = 23)







```
